```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from scipy.optimize import leastsq
##定义原函数
def real_func(x):
    return np.sin(np.pi*2*x)

##设置拟合多项式
def func(p,x):
    f=np.poly1d(p)
    return f(x)

##定义残差
def residuals_func(p,x,y):
    return func(p,x)-y


x_sample=np.linspace(0,1,10)
x_population=np.linspace(0,1,1000)
y=real_func(x_sample)
y_noise=[y_+np.random.normal(0,0.1) for y_ in y]
def p_random(x):
    return np.random.rand(x+1)

def fitting(N):
    '''N是多项式的次数'''
    sq=leastsq(residuals_func,p_random(N),args=(x_sample,y_noise))
    
    ##可视化
    plt.plot(x_population,real_func(x_population),label='real_func')
    plt.plot(x_sample,func(sq[0],x_sample),label='fitting curve')
    plt.plot(x_sample,y_noise,'b^',label='sample')
    plt.legend()
    return sq[0]
```

```python
fitting(9)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from scipy.optimize import leastsq

def real_func(x):
    return np.sin(2*np.pi*x)

def func(p,x):
    f= np.poly1d(p)
    return f(x)

def resduals_func(p,x,y):
    return func(p,x)-y

x_sample=np.linspace(0,1,10)
x_population=np.linspace(0,1,1000)
y=real_func(x_sample)
y_noise=[y_+np.random.normal(0,0.1) for y_ in y]

def fitting(N):
    p=np.random.rand(N+1)
    sq=leastsq(resduals_func,p,(x_sample,y_noise))
    
    plt.plot(x_sample,y,'b^',label='noise')
    plt.plot(x_sample,func(sq[0],x_sample),label='fitting curve')
    plt.plot(x_population,real_func(x_population),label='real_func')
    print(sq[0])
    plt.legend()

fitting(9)
```
